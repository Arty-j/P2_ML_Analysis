```python
# Import
import pandas as pd
from pathlib import Path
```

```python
coin = "BTC"
```

```python
# Read in CSV file with Exchange-Coin data
df = pd.read_csv(Path(f"./USD_pairs/Coinbase/daily/Coinbase_{coin}USD_dailydata.csv"), parse_dates= True, infer_datetime_format = True)
df.head()
```

```python
# reformat the dataframe to include OHLCV data for years 2021-2022
df.insert(loc=0, column = "Exchange", value = "Coinbase") 
df.insert(loc=1, column = "Symbol_Pair", value = f"{coin}/USD") 
df.rename(columns = {"unix":"Unix", "low": "Low", "high": "High",
                        "open": "Open", "close": "Close", "volume": f"Volume_in_{coin}", "date": "Datetime",
                        "vol_fiat": "Volume_USD"},
              inplace = True)

df = df.drop(["Unix"], axis=1)
df = df.set_index("Datetime")
```

```python
df.index = pd.to_datetime(df.index)
```
