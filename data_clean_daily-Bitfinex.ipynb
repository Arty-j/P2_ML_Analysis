```python
# Import
import pandas as pd
from pathlib import Path
```

```python
# Read in CSV file with Exchange-Coin data
#manually changed between Bitfinex, Gemini and all coins

# to get hourly data changed foler name from daily to hourly and changed csv file name to download name
df = pd.read_csv(Path("./USD_pairs/Bitfinex/daily/Bitfinex_BTCUSD_d.csv"), parse_dates= True, infer_datetime_format = True)
df.head()
```

```python
# reformat the dataframe to include OHLCV data for years 2021-2022
df = df.reset_index()
df.rename(columns = {"level_0":"Unix", "level_1": "Datetime", "level_2": "Symbol_Pair",
                        "level_3": "Open", "level_4": "High", "level_5": "Low", "level_6": "Close",
                        "level_7": "Volume_USD", "https://www.CryptoDataDownload.com": "Volume_in_BTC"},
              inplace = True)
df = df.drop([0,1], axis=0)
df = df.drop(["Unix"], axis=1)
df = df.set_index("Datetime")
```

```python
df.index = pd.to_datetime(df.index)
df.insert(loc=0, column = "Exchange", value= "Bitfinex") 
```

```python
# change data type to string and float
df = df.astype({'Symbol_Pair': str})
columns = ['Open', 'High', 'Low', 'Close', 'Volume_USD', 'Volume_in_BTC']
for c in columns:
    df = df.astype({c: float})
```
