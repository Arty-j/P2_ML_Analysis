```python
import datetime
import time
import json
import requests
import pandas as pd
from pandas import json_normalize
```

```python
url = f"https://api.coingecko.com/api/v3/exchanges/binance/tickers?coin_ids=usd_coin"
response = requests.get(url)
data = response.json()
df = json_normalize(data['tickers'][0])
df.head()      
```

```python
df["timestamp"] = pd.to_datetime(
    df["timestamp"],
    infer_datetime_format =True,
    utc = True
)
df.info()
```

```python
df1 = df[['base',
          'target', 'last', 'volume', 'bid_ask_spread_percentage',
          'timestamp', 'market.name', 'market.identifier', 'market.has_trading_incentive',
          'converted_last.usd', 'converted_volume.usd']]
df1
```
