```python
# Import
import pandas as pd
from pathlib import Path
```

```python
# Read in CSV histoical data file from Tether_pairs which was downloaded from CryptoDataDownload.com

# ID coin and exchange variables
coin = "XRP"
exchange = "Binance"

df = pd.read_csv(Path(f"./USD_pairs/{exchange}/{exchange}_{coin}USDT_d.csv"), parse_dates= True, infer_datetime_format = True)
df.head()
```

```python
df = df.reset_index()
df
```

```python
# reformat the dataframe to set column names, drop unnecessary rows and columns, move Datetime column to index

df.rename(columns = {"level_0":"Unix", "level_1": "Datetime", "level_2": "Symbol_Pair",
                        "level_3": "Open", "level_4": "High", "level_5": "Low", "level_6": "Close",
                        "level_7": f"Volume_in_{coin}", "level_8": "Volume_USDT", "https://www.CryptoDataDownload.com": "Tradecount"},
              inplace = True)
df = df.drop([0,1], axis=0)
df = df.drop(["Unix"], axis=1)
df = df.set_index("Datetime")
```

```python
# change index to dtype datetime
df.index = pd.to_datetime(df.index)

# Insert column for exchange name
df.insert(loc=0, column = "Exchange", value= f"{exchange}") 
```

```python
display(df.head(5))
df.tail()
```

```python
# check data types
df.info()
```

```python
# change data type to float
columns = ['Open', 'High', 'Low', 'Close', f'Volume_in_{coin}', 'Volume_USDT', 'Tradecount']
for c in columns:
    df = df.astype({c: float})
```

```python
df.info()
```

```python
# check for null values
df.isna().sum().sum()
```

```python
# slice data from 2021 & 2022
df = df.loc['2021-01-01':'2022-12-27']
display(df.head())
df.tail()
```

```python
# make a copy of new dataframe and export to Data file for use in ML models
df_export = df.copy()
df_export.to_csv(f'./Data/{exchange}/21-22{exchange}_{coin}_d.csv')
```
