```python
# Import
import pandas as pd
from pathlib import Path
import numpy as np
import os
import glob
```

```python
# Read in CSV file with Exchange-Coin data
# change folder names and echange name to calculate all DF for Bitfinex, Gemini, Coinbase

all_files = glob.glob(os.path.join("./Data/Coinbase/21-22Coinbase_*.csv"))                      
li = []
                      
for filename in all_files:
    df = pd.read_csv(filename, index_col = "Datetime", parse_dates=True, infer_datetime_format = True, header=0)
    li.append(df)
    
all_df = pd.concat(li, axis=0) 
all_df.head()
```

```python
all_df.shape
```

```python
df_export = all_df.copy()
df_export.to_csv('./Data/Coinbase/all_21-22Coinbase_d.csv')
```
